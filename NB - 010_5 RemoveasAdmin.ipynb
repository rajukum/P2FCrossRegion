```python
import sempy_labs as labs
import sempy_labs.admin as labs_admin
from pyspark.sql import Row
import datetime
```

```python
def log_to_table(level, message):
    now = datetime.datetime.utcnow().isoformat()
    log_df = spark.createDataFrame([Row(timestamp=now, level=level, message=message)])
    log_df.write.format("delta").mode("append").saveAsTable("workspaceRemoveUser")
```

```python
current_user = mssparkutils.env.getUserName()
workspaces_df = spark.sql(f"Select DISTINCT Workspace_Id from models_discovery_status")
log_to_table("INFO", f"Removing user {current_user} from workspaces started.")
for row in workspaces_df.collect():
    ws_id = row['Workspace_Id']
    try:
      labs_admin.delete_user_from_workspace(workspace=ws_id, user=current_user)
      print(f" -User removed from workspace ID.'{ws_id}'.")
      log_to_table("INFO", f"User removed from workspace ID. {ws_id}.")
    except Exception as e:
        print(f"  ‚ùå FAILED: An error occurred while removing user from workspace {ws_id}.")
        print(f"    Error details: {e}")
        log_to_table("ERROR", f"An error occurred while removing user from  workspace {ws_id} error: {str(e)}")
log_to_table("INFO", f"Removing user {current_user} from workspaces Completed.")
```
